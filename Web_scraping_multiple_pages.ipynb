```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd

# Set up the Selenium driver
driver = webdriver.Chrome()

# Navigate to the webpage
driver.get("https://myanimelist.net/topmanga.php?type=manga")

# Wait for the table containing the top manga to load
table = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CLASS_NAME, "top-ranking-table")))

# Create a list to store it into the DataFrame
list_comics = []

# Loop through each row and extract the title and score
while True:
    try:
        # Find the rows of the table - containing img, title, score in <td> of HTML (INSTEAD OF TITLE)
        rows = driver.find_elements(By.CLASS_NAME, "ranking-list")
        for row in rows:
            title = row.find_element(By.CLASS_NAME, "hoverinfo_trigger.fs14.fw-b").text
            des = row.find_element(By.CLASS_NAME, "information.di-ib.mt4").text.replace("\n", " ")
            score = row.find_element(By.XPATH, ".//td[@class='score ac fs14']").text.strip()
            url = row.find_element(By.TAG_NAME, "a").get_attribute("href")
            img = row.find_element(By.TAG_NAME, "img").get_attribute("data-src")

            # CREATE HEADER WITH VALUE WE GET FROM ABOVE FOR EACH "title" , "des" , and "score"
            comic_header = {"Title" : title, "Description" : des, "Score" : score, "URL" : url, "img_url" : img}
            list_comics.append(comic_header)
            
        # WE NEED TO CLICK TO THE NEXT PAGE IN THE WEB PAGE TO GET MORE INFORMATION
        next = driver.find_element(By.CLASS_NAME, "link-blue-box.next")
        next.click()
        
        # FINALLY, WE PUT IT IN THE DATAFRAME TO DISPLAY IT
        df = pd.DataFrame(list_comics)
        print(df)  
    except NoSuchElementException:
        break    
  
        
# Close the Selenium driver

```

```python
df.to_excel("full-top_manga_list.xlsx")
```
